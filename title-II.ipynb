```python
import numpy as np
import pandas as pd
import seaborn as sns
#Code to read files
Title2 = pd.read_excel("D:\\T2.xlsx")
Alt = pd.read_excel("D:\\Alt.xlsx")
UG = pd.read_excel("D:\\UG.xlsx")
```

```python
#Code to consider only Reporting group 2 from the table
Title2=Title2[Title2['Reporting Group']==2.0]

#Code to convert Float to Int type
Title2['919'] = pd.to_numeric(Title2['919']).round(0).astype(int)
```

```python
#Code to compare 919s from two different files and change reporting group based on a date
for i in range(len(Alt['ID'])):
    for j in range(len(Title2['919'])):
        if(Alt['ID'][i]==Title2['919'][j]):
            if(Alt['ALTCOMPDATE'][i]<Complete_date):
                Title2['Reporting Group'][j]=3
```

```python
#Code to compare Alt-certs and Title2 files to make changes on Title2
pd.to_numeric(Alt['ID'])
for j in range(len(Alt['ID'])):
    if(Alt['ID'][j] not in Title2['919'].values):
        Title2.loc[Title2.index.max() + 1] = '.'
        Title2['919'].loc[Title2.index.max()]=Alt['ID'][j]
        Title2['Last Name'].loc[Title2.index.max()]=Alt['LAST_NAME'][j]
        Title2['First Name'].loc[Title2.index.max()]=Alt['FIRST_NAME'][j]
        Title2['Program Type'].loc[Title2.index.max()]='A'
        if Alt['GENDER_DESC'][j]=="Female":
            Title2['Gender'].loc[Title2.index.max()]="F"
        if Alt['GENDER_DESC'][j]=="Male":
            Title2['Gender'].loc[Title2.index.max()]="M"
        Title2['Race'].loc[Title2.index.max()]=Alt['ETHNICITY'][j]
        Title2['GPA 2018-2019 (report median)'].loc[Title2.index.max()]=Alt['GPA'][j]
        if(Alt['ALTCOMPDATE'][j]<complete_date):
            Title2['Reporting Group'].loc[Title2.index.max()]=3
        else:
            Title2['Reporting Group'].loc[Title2.index.max()]=2
```

```python
#Code to define a function to remove Delimiters
delimiters = [",", ".", "!", "?", "/", "&", "-", ":", ";", "@", "'", "...","(",")"]
def remove_delimiters (s):
    new_s = str(s)
    for i in delimiters: #replace each delimiter in turn with a space
        new_s = new_s.replace(i, '')
    return ''.join(new_s.split())
```

```python
#Code to Map all the columns with remove Delimiters function
abc['Major 1']=abc['Major 1'].map(remove_delimiters)
abc['Major 2']=abc['Major 2'].map(remove_delimiters)
abc['MAJOR 3']=abc['MAJOR 3'].map(remove_delimiters)
abc['MAJOR 4']=abc['MAJOR 4'].map(remove_delimiters)
abc['MINOR 1']=abc['MINOR 1'].map(remove_delimiters)
abc['MINOR 2']=abc['MINOR 2'].map(remove_delimiters)
abc['MINOR 3']=abc['MINOR 3'].map(remove_delimiters)
abc['MINOR 4']=abc['MINOR 4'].map(remove_delimiters)
```
